```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g=9.8
m=40
m2=20
M=30
```

```python
def primera(x,e=1E+3):
    r1=np.sqrt((e-g*(m-m2)*x)*2*(m+m2))
    return r1
```

```python
def segunda(x,e=1E+3):
    r1=np.sqrt((e-g*(m-m2)*x)*2*(m+m2+M))
    return r1
```

```python
def H0(p0,z0):
    t1=pow(p0,2)/(2*m)
    t2=g*m*z0
    return t1+t2
```

```python
def pz(x,p0=5,z0=50,dp0=7,dz0=25):
    h0=H0(p0+dp0,z0+dz0)
    t1=h0-m*g*x
    t2=2*m
    return np.sqrt(t1*t2)
```

```python
xMin, xMax, N = 0.,80, 200
x = np.linspace(xMin, xMax, N)
h = (xMax-xMin)/float(N-1)
```

```python
F1=primera(x)
```

```python
F2=segunda(x)
```

```python
plt.plot(x,F1,label='Sin considerar polea')
plt.plot(x,F2,label='Considerando polea')
plt.title('Espacio de Fase')
plt.xlabel('x[m]')
plt.ylabel('Px[m]')
plt.legend()
plt.savefig('../../Images/Espacio de fase')
```

```python
G1=pz(x,dp0=0,dz0=0)
```

```python
G2=pz(x,dp0=0)
```

```python
G3=pz(x,dz0=0)
```

```python
G4=pz(x)
```

```python
plt.plot(x,G1)
plt.plot(x,G2)
plt.plot(x,G3)
plt.plot(x,G4)
```
