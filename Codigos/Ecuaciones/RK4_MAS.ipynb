```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.animation as anim
from tqdm import tqdm
```

```python
def dv(x,m=3,k=5):
    return -20*x/m

def dx(v):
    return v

def cua(t):
    return t**2
```

```python
t=np.linspace(0,10,20000)
T=len(t)
h=(t[-1]-t[0])/T
print(h)
```

```python
V=np.zeros(T)
X=np.zeros(T)
V[0]=1
X[0]=0
```

```python
def Integracion(f1,f2):
    for i in tqdm(range(1,T)):
        #valores para cada iteracion
        vi=V[i-1]
        xi=X[i-1]
        
        #solucion en x
        k1v=f1(xi)
        k1x=f2(vi)
        k2v=f1(xi+h*k1v/2)
        k2x=f2(vi+h*k1x/2)
        k3v=f1(xi+h*k2v/2)
        k3x=f2(vi+h*k2x/2)
        k4v=f1(xi+h*k3v)
        k4x=f2(vi+h*k3x)
        
        V[i]=vi+h/6*(k1v+2*k2v+2*k3v+k4v)
        X[i]=xi+ h/6* (k1x+2*k2x+2*k3x+k4x)

        
```

```python
Integracion(dv,dx)
```

```python
plt.plot(t,V)
plt.plot(t,X)
```
