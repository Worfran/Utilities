```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi=np.pi
```

```python
def bn(n):
    factor1=-(pi)/n * (-1)**n
    factor2=-(6)/pow(n,3) * (-1)**n
    return factor1+factor2
```

```python
def fourier(x,n):
    s=0
    for i in range(1,n+1):
        s+=bn(i)*np.sin(x*i)
    return s
        
```

```python
x=np.linspace(-pi,pi,100)
```

```python
s=[2,5,20]
results=[]
```

```python
for i in s:
    y=fourier(x,i)
    results.append(y)
    
```

```python
j=0
while j < len(results):
    plt.plot(x,results[j],label="n={}".format(s[j]))
    j+=1

plt.legend()
```
